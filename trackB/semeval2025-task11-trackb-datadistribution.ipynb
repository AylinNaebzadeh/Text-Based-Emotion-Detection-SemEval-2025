```python
def plot_label_distribution_combined(df, label_columns):
    """
    Plots the distribution of values for all labels in a single grouped bar chart.
    """
    # Prepare the data
    value_counts = {label: df[label].value_counts(sort=False) for label in label_columns}
    categories = sorted(set().union(*[counts.index for counts in value_counts.values()]))  # Unique intensity values
    bar_width = 0.15  # Width of each bar
    x = np.arange(len(categories))  # X positions for the categories

    # Plot
    plt.figure(figsize=(10, 6))
    for i, label in enumerate(label_columns):
        counts = value_counts[label].reindex(categories, fill_value=0)  # Fill missing intensity values with 0
        plt.bar(x + i * bar_width, counts, width=bar_width, label=label)

    # Customize the plot
    plt.title('Label Distribution Across Intensities')
    plt.xlabel('Intensity')
    plt.ylabel('Count')
    plt.xticks(x + (len(label_columns) - 1) * bar_width / 2, categories)  # Center the ticks
    plt.legend(title='Labels')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
label_columns = ['anger', 'disgust', 'fear', 'joy', 'sadness', 'surprise']
plot_label_distribution_combined(train_rus, label_columns)
```

```python
plot_label_distribution_combined(train_ukr, label_columns)
```

```python
plot_label_distribution_combined(train_ron, label_columns)
```

```python
plot_label_distribution_combined(train_ptbr, label_columns)
```
