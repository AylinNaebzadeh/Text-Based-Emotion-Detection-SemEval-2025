```python
# Define a function to translate text
def translate_text_to_english(text):
    try:
        if pd.isna(text):  # Skip NaN values
            return text
        return GoogleTranslator(source='auto', target='en').translate(text) 
    except Exception as e:
        print(f"Error translating text: {text}. Error: {e}")
        return text
```

```python
# Define file paths and names
file_mappings = {
    "rus": "/kaggle/input/semeval-2025-task-11-emotion/public_data/train/track_b/rus.csv",
    "deu": "/kaggle/input/semeval-2025-task-11-emotion/public_data/train/track_b/deu.csv",
    "ptbr": "/kaggle/input/semeval-2025-task-11-emotion/public_data/train/track_b/ptbr.csv",
    # "rus_dev": "/kaggle/input/semeval-2025-task-11-emotion/public_data/dev/track_b/rus_b.csv",
    # "deu_dev": "/kaggle/input/semeval-2025-task-11-emotion/public_data/dev/track_b/deu_b.csv",
    # "ptbr_dev": "/kaggle/input/semeval-2025-task-11-emotion/public_data/dev/track_b/ptbr_b.csv",
}
```

```python
# Process each file and translate text
for lang, file_path in file_mappings.items():
    print(f"Processing file: {file_path}")
    df = pd.read_csv(file_path)
    
    if 'text' in df.columns:
        df['text_to_en'] = df['text'].progress_apply(translate_text_to_english)
        output_path = f"{lang}_to_eng.csv"
        df.to_csv(output_path, index=False)
        print(f"Saved translated file as: {output_path}")
    else:
        print(f"Column 'text' not found in {file_path}")
```
